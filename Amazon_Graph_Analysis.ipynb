# Amazon Graph Analysis with NetworkX
This notebook demonstrates how to:
- Create a graph from an edge list (Amazon dataset)
- Compute **Degree Centrality** and **Betweenness Centrality**
- Visualize the graph with NetworkX and Matplotlib

```python
# Import necessary libraries
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

## 1. Load the Dataset (Amazon Graph)
The dataset consists of edges representing relationships between products.

```python
# Define the dataset as an edge list (Amazon Graph Sample)
edges = [
    (161797, 36000), (161797, 521221),
    (272396, 38365), (272396, 476945), (272396, 477105), (272396, 513227),
    (215057, 38365),
    (366610, 55488), (366610, 246481), (366610, 377917),
    (366610, 470072), (366610, 483197), (366610, 483741),
    (399385, 38365), (399385, 175102), (399385, 431174), (399385, 458358),
    (516125, 24139), (516125, 62917), (516125, 175102)
]

# Create a Graph
G = nx.Graph()
G.add_edges_from(edges)
```

## 2. Compute Centrality Measures
**Degree Centrality:** Measures node importance based on connections.
**Betweenness Centrality:** Measures influence based on shortest paths.

```python
# Compute Degree Centrality
degree_centrality = nx.degree_centrality(G)

# Compute Betweenness Centrality
betweenness_centrality = nx.betweenness_centrality(G)

# Print the top 5 nodes by Degree Centrality
sorted_degree = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
print("Top 5 Nodes by Degree Centrality:")
for node, centrality in sorted_degree:
    print(f"Node {node}: {centrality:.4f}")

# Print the top 5 nodes by Betweenness Centrality
sorted_betweenness = sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:5]
print("\nTop 5 Nodes by Betweenness Centrality:")
for node, centrality in sorted_betweenness:
    print(f"Node {node}: {centrality:.4f}")
```

## 3. Visualize the Graph
Nodes with higher centrality will be larger in size.

```python
# Draw the graph
plt.figure(figsize=(10, 6))

# Node size based on Degree Centrality
node_size = [5000 * degree_centrality[node] for node in G.nodes()]

# Draw network
nx.draw(G, with_labels=True, node_size=node_size, node_color='skyblue', edge_color='gray')
plt.title("Amazon Graph Visualization with Degree-Based Node Sizing")
plt.show()
```
